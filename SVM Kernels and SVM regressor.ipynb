```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mpl_toolkits import mplot3d
from matplotlib.axes._axes import _log as matlotlib_axes_logger
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=100,noise=0.1,factor=0.1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
classifier = SVC(kernel="linear") # let's try using a linear kernel to separate a non-linearly separable data
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_true=y_test,y_pred=y_pred)
```

```python
# we get a very bad accuracy = 30 %
```

```python
# let's use a different kernel
```

```python
new_classifier = SVC(kernel='rbf') # using rbf kernel
```

```python
new_classifier.fit(X_train,y_train)
```

```python
y_pred_rbf = new_classifier.predict(X_test)
y_pred_rbf
```

```python
y_test
```

```python
accuracy_score(y_true=y_test,y_pred=y_pred_rbf)
```

```python
# we get 100 % accuracy
```

```python
# let's try using the polynomial kernel
```

```python
poly_classifier = SVC(kernel='poly')
```

```python
poly_classifier.fit(X_train,y_train)
```

```python
y_pred3 = poly_classifier.predict(X_test)
y_pred3
```

```python
accuracy_score(y_true=y_test,y_pred=y_pred3)
```

```python
classifier4 = SVC(kernel='poly',degree=2)
```

```python
classifier4.fit(X_train,y_train)
```

```python
y_pred4 = classifier4.predict(X_test)
y_pred4
```

```python
accuracy_score(y_true=y_test,y_pred=y_pred4)
```

```python
def plot_3d_plot(X,y):
    r = np.exp(-(X**2).sum(1))
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:,0],X[:,1],r,c=y,s=100,cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
```

```python
plot_3d_plot(X,y)
```
